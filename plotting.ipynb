**Task 1**

```python
#preparing the required data by using linspace to get 1000 values in range 0 to 10
data=np.linspace(0,10,1000)
```

```python
#Setting axes to plot the data
fig=plt.figure()
myAxes=plt.axes()
```

```python
myAxes.plot(data,np.sin(data));
plt.title("1000 evenly spaced numbers")
plt.xlabel("X value")
plt.ylabel("Sin(x)")
plt.plot(data,np.sin(data-0),':',color='b')
plt.plot(data,np.sin(data-1),'--',color='tab:orange')
plt.plot(data,np.sin(data+1),color='r')
```
